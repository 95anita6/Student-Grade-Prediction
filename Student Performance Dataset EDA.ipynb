```python
# seperate numerical and categorical features to check the relation:
num_feat = sp.select_dtypes(include=[np.number])
cat_feat = sp.select_dtypes(exclude=[np.number])
```

```python
num_feat
```

```python
cat_feat.head()
```

```python
# Using correlation coefficient to find relation among the numeric features:
corr = num_feat.corr()

plt.figure(figsize=(14,14))
sns.heatmap(corr, annot=True, fmt= '.2f',annot_kws={'size': 15}, cmap= 'coolwarm')
```

```python
# There is strong correlation between the grades G1,G2 and final G3 of which G2 has a stronger relation as compared to G1.
# There also seems to be strong correlation among weekday and weekend alcohol consumption but these features indiviually have negative correlation with the target variable.
# There also seems to be correlation between Father's and mother's education.
# Most of the other numeric variables seem to be in negative correlation with the target.
```

```python
from scipy import stats
cat = cat_feat
def anova(frame):
    anv = pd.DataFrame()
    anv['features'] = cat.columns
    pvals = []
    for c in cat:
        samples = []
        for cls in frame[c].unique():
            s = frame[frame[c] == cls]['G3'].values
            samples.append(s)
        pval = stats.f_oneway(*samples)[1]
        pvals.append(pval)
    anv['pval'] = pvals
    return anv.sort_values('pval')
```

```python
cat_feat['G3'] = sp.G3.values
k = anova(cat_feat) 
k['disparity'] = np.log(1./k['pval'].values) 
sns.barplot(data=k, x = 'features', y='disparity') 
plt.xticks(rotation=90) 
plt
```
